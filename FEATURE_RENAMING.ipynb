```python
import pandas as pd

# Load your CSV
df = pd.read_csv("f.i v2 complete data/FI V2 COMPLETE FEATURE DATASET/HANDCRAFTED/audio_hc_features.csv")  # Replace with your actual path


# eGeMAPS v02 feature names (25 features)
egemaps_names = [
    "F0semitoneFrom27.5Hz", "jitterLocal", "shimmerLocal", "HNRdBACF",
    "logRelF0-H1-H2", "logRelF0-H1-A3", "F1frequency", "F1bandwidth",
    "F2frequency", "F2bandwidth", "F3frequency", "F3bandwidth",
    "alphaRatio", "hammarbergIndex", "slope0-500", "slope500-1500",
    "spectralFlux", "mfcc1", "mfcc2", "mfcc3", "mfcc4",
    "loudness", "voiceProb", "logRelF0-H1-A1", "logRelF0-H1-A2"
]

# Rename only the hc_0 to hc_24 columns
# Find their positions in the DataFrame
hc_columns = [col for col in df.columns if col.startswith("Audio_hc_")]

# Ensure correct count
assert len(hc_columns) == 25, f"Expected 25 hc_ columns, found {len(hc_columns)}."

# Create mapping from hc_0...hc_24 to eGeMAPS names
rename_dict = dict(zip(hc_columns, egemaps_names))

# Apply the renaming
df.rename(columns=rename_dict, inplace=True)

# Save the cleaned CSV
df.to_csv("f.i v2 complete data/FI V2 COMPLETE FEATURE DATASET/HANDCRAFTED/renamed_audio_hc_features.csv", index=False)

print("Renaming complete. File saved as 'renamed_full_features.csv'.")

```

```python
import pandas as pd

# === Step 1: Extract LIWC Category Names from LIWC2007.txt ===
liwc_file = "f.i v2 complete data/FI V2 COMPLETE FEATURE DATASET/HANDCRAFTED/LIWC2007.txt"
liwc_categories = []

with open(liwc_file, 'r') as f:
    lines = f.readlines()

start = lines.index('%\n') + 1
end = lines[start:].index('%\n') + start

for line in lines[start:end]:
    parts = line.strip().split()
    if len(parts) >= 2:
        liwc_categories.append(parts[1])

# Validate count
if len(liwc_categories) != 64:
    raise ValueError(f"Expected 64 LIWC categories, found {len(liwc_categories)}.")

# === Step 2: Load your feature CSV ===
df = pd.read_csv("f.i v2 complete data/FI V2 COMPLETE FEATURE DATASET/HANDCRAFTED/text_hc_features.csv")  # Replace with actual file

# Find index where Text_hc features start
hc_start = df.columns.get_loc("Text_hc_0")
hc_end = hc_start + 64

# Rename 'Text_hc_0' ... 'Text_hc_63' to LIWC names
new_columns = df.columns.tolist()
new_columns[hc_start:hc_end] = liwc_categories
df.columns = new_columns

# === Step 3: Save the renamed file ===
df.to_csv("f.i v2 complete data/FI V2 COMPLETE FEATURE DATASET/HANDCRAFTED/renamed_text_hc_features.csv", index=False)
print("Successfully renamed LIWC columns and saved to 'renamed_text_hc_features.csv'.")
```

```python
frames_per_segment = 10
angle_names = [
    "Angle_Between_Eyebrows",
    "Angle_Between_Eyes_Centers",
    "Angle_Between_Lip_Corners"
]
distance_names = [f"Distance_Landmark_{i}_Pair" for i in range(36)]
landmark_coord_names = [f"Landmark_{i}_X" for i in range(27)] + [f"Landmark_{i}_Y" for i in range(27)]

frame_features = angle_names + distance_names + landmark_coord_names

descriptive_feature_names = []
for frame in range(frames_per_segment):
    for feature in frame_features:
        descriptive_feature_names.append(f"Frame_{frame}_{feature}")

# Add the 6 extra features at the end
extra_features = [f"Global_Feature_{i}" for i in range(6)]
descriptive_feature_names.extend(extra_features)

# Optionally: save to file
with open("f.i v2 complete data/FI V2 COMPLETE FEATURE DATASET/HANDCRAFTED/video_hc_descriptive_names.txt", "w") as f:
    f.write("\n".join(descriptive_feature_names))
```
