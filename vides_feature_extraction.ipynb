```python
def preprocess_individual_audio(file_path, n_mels=25, sr=16000, duration=2):
    """
    Preprocess an individual's audio file into a fixed-length mel spectrogram tensor.

    :param file_path: Path to the audio file
    :param n_mels: Number of mel bands (must match LSTM input size)
    :param sr: Sampling rate
    :param duration: Fixed duration (in seconds) for feature extraction
    :return: PyTorch tensor of shape (1, time_steps, n_mels)
    """
    import librosa
    import numpy as np
    import torch

    # Load the audio as mono
    y, sr = librosa.load(file_path, sr=sr, mono=True)

    # Normalize audio length by padding/trimming
    max_length = int(sr * duration)  # Fixed duration in samples
    y = librosa.util.fix_length(y, size=max_length)

    # Compute mel spectrogram
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels)

    # Convert power spectrogram to dB scale
    mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # Return tensor with shape (1, time_steps, n_mels)
    return torch.tensor(mel_spectrogram_db.T).unsqueeze(0).float()
```

```python
import cv2
import numpy as np

def segment_video(video_path, segment_duration=2, overlap=1, fps=5):
    """
    Segments a video into overlapping chunks.

    :param video_path: Path to the video file
    :param segment_duration: Duration of each segment (seconds)
    :param overlap: Overlap between segments (seconds)
    :param fps: Frames per second to sample
    :return: List of video segments as numpy arrays
    """
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_fps = int(cap.get(cv2.CAP_PROP_FPS))
    downsample_rate = max(1, video_fps // fps)
    
    # Read frames and downsample
    frames = []
    for i in range(total_frames):
        ret, frame = cap.read()
        if not ret:
            break
        if i % downsample_rate == 0:  # Downsample
            frames.append(frame)
    cap.release()

    # Create overlapping segments
    frame_count = len(frames)
    segment_frames = int(segment_duration * fps)
    step_frames = int((segment_duration - overlap) * fps)

    segments = []
    for start in range(0, frame_count - segment_frames + 1, step_frames):
        segments.append(frames[start:start + segment_frames])

    return np.array(segments)  # Shape: (num_segments, segment_frames, H, W, C)
```

```python
import torch
import torchvision.models as models
from torchvision import transforms
from PIL import Image
import cv2

# Load a pre-trained ResNet50 model
emotion_model = models.resnet50(pretrained=True)
emotion_model.fc = torch.nn.Linear(emotion_model.fc.in_features, 8)  # Assume 8 emotion classes
emotion_model.eval()

# Transformation pipeline for input images
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def extract_deep_emotional_features(frame):
    """
    Extract deep emotional features from a single video frame.

    :param frame: Video frame (H, W, C) as a numpy array
    :return: Deep emotional features as a PyTorch tensor
    """
    # Convert the frame to PIL Image
    image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    # Preprocess the image
    input_tensor = preprocess(image).unsqueeze(0)  # Add batch dimension

    # Extract features
    with torch.no_grad():
        features = emotion_model(input_tensor)
    return features.squeeze().numpy()  # Return as a numpy array
```

```python
def extract_features_from_segment(segment):
    """
    Extract features from a single video segment.

    :param segment: A video segment (10 frames, H, W, C)
    :return: Deep emotional features and geometric features
    """
    # Initialize Face Mesh model
    face_mesh = FaceMesh(static_image_mode=True)

    deep_features = []
    geometric_features = []

    for frame in segment:
        # Extract face landmarks
        results = face_mesh.process(frame)

        if results.multi_face_landmarks:
            landmarks = results.multi_face_landmarks[0]
            
            # Extract deep emotional features using pre-trained model
            deep_feat = extract_deep_emotional_features(frame)  # Shape: (8,)
            deep_features.append(deep_feat)

            # Extract geometric features (example logic)
            geo_feat = extract_geometric_features(landmarks)
            geometric_features.append(geo_feat)
        else:
            # Handle case where no face is detected
            deep_features.append(np.zeros(8))  # Update size based on model output
            geometric_features.append(np.zeros(115))

    return np.array(deep_features), np.array(geometric_features)
```

```python
# def segment_video(video_path, segment_duration=2, overlap=1, fps=5):
#     """
#     Segments a video into overlapping chunks.

#     :param video_path: Path to the video file
#     :param segment_duration: Duration of each segment (seconds)
#     :param overlap: Overlap between segments (seconds)
#     :param fps: Frames per second to sample
#     :return: List of video segments as numpy arrays
#     """
#     cap = cv2.VideoCapture(video_path)
#     total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
#     video_fps = int(cap.get(cv2.CAP_PROP_FPS))
#     downsample_rate = max(1, video_fps // fps)
    
#     frames = []
#     for i in range(total_frames):
#         ret, frame = cap.read()
#         if not ret:
#             break
#         if i % downsample_rate == 0:  # Downsample
#             frames.append(frame)
#     cap.release()

#     frame_count = len(frames)
#     segment_frames = int(segment_duration * fps)
#     step_frames = int((segment_duration - overlap) * fps)

#     segments = []
#     for start in range(0, frame_count - segment_frames + 1, step_frames):
#         segments.append(frames[start:start + segment_frames])

#     return np.array(segments)


# def segment_video(video_path, segment_duration=2, fps=1):
#     """
#     Segments a video into non-overlapping chunks with skipped frames.

#     :param video_path: Path to the video file
#     :param segment_duration: Duration of each segment (in seconds)
#     :param fps: Frames per second to sample
#     :return: List of video segments as numpy arrays
#     """
#     cap = cv2.VideoCapture(video_path)
#     total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
#     video_fps = int(cap.get(cv2.CAP_PROP_FPS))
#     downsample_rate = max(1, video_fps // fps)
    
#     # Extract frames at 1 FPS
#     frames = []
#     for i in range(total_frames):
#         ret, frame = cap.read()
#         if not ret:
#             break
#         if i % downsample_rate == 0:  # Downsample
#             frames.append(frame)
#     cap.release()

#     # Ensure the correct frame sampling
#     segment_frames = segment_duration * fps  # 2 frames for 2 seconds
#     segments = []
#     start = 0

#     while start + segment_frames <= len(frames):
#         segments.append(frames[start:start + segment_frames])
#         start += segment_frames + 1  # Skip 1 frame after each segment

#     return np.array(segments)





# def segment_video(video_path, segment_duration=2, fps=1):
#     """
#     Segments a video into non-overlapping chunks with skipped frames.

#     :param video_path: Path to the video file
#     :param segment_duration: Duration of each segment (in seconds)
#     :param fps: Frames per second to sample
#     :return: List of video segments as numpy arrays
#     """
#     cap = cv2.VideoCapture(video_path)
#     video_fps = int(cap.get(cv2.CAP_PROP_FPS))
#     downsample_rate = max(1, video_fps // fps)
#     segment_frames = segment_duration * fps  # Frames per segment
    
#     segments = []
#     frame_buffer = []
#     frame_index = 0

#     while True:
#         ret, frame = cap.read()
#         if not ret:
#             break

#         # Downsample to the desired FPS
#         if frame_index % downsample_rate == 0:
#             frame_buffer.append(frame)

#         # If enough frames are collected for a segment
#         if len(frame_buffer) == segment_frames:
#             segments.append(frame_buffer)
#             frame_buffer = []  # Clear buffer for next segment

#             # Skip 1 frame (1 second at 1 FPS)
#             for _ in range(fps): 
#                 if not cap.read()[0]:
#                     break

#         frame_index += 1

#     cap.release()
#     return np.array(segments)





def segment_video(video_path, segment_duration=2, fps=1):
    """
    Segments a video into non-overlapping chunks with skipped frames.

    :param video_path: Path to the video file
    :param segment_duration: Duration of each segment (in seconds)
    :param fps: Frames per second to sample
    :return: List of video segments as numpy arrays
    """
    cap = cv2.VideoCapture(video_path)
    video_fps = int(cap.get(cv2.CAP_PROP_FPS))
    downsample_rate = max(1, video_fps // fps)

    # Extract frames at 1 FPS
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if int(cap.get(cv2.CAP_PROP_POS_FRAMES)) % downsample_rate == 0:
            frames.append(frame)
    cap.release()

    # Ensure the correct frame sampling
    segment_frames = segment_duration * fps  # 2 frames for 2 seconds
    segments = []
    frame_indices = [1, 4, 7, 10, 13]  # Start frames for the 5 segments

    for start in frame_indices:
        if start + segment_frames <= len(frames):
            segments.append(frames[start:start + segment_frames])

    return np.array(segments)
```

```python
import cv2
import numpy as np

def segment_video_fixed(video_path, fps=5, segment_duration=2, step=1):
    """
    Segment a video into overlapping segments.

    :param video_path: Path to the video file
    :param fps: Frames per second for downsampling
    :param segment_duration: Duration of each segment in seconds
    :param step: Step size for overlapping segments in seconds
    :return: List of segments (each segment is a numpy array of shape [frames, height, width, channels])
    """
    cap = cv2.VideoCapture(video_path)
    all_frames = []
    
    # Read frames
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        all_frames.append(frame)
    cap.release()
    
    # Downsample frames to target FPS
    target_frames = int(fps * segment_duration)
    total_frames = len(all_frames)
    segments = []
    
    for start in range(0, total_frames, fps * step):
        end = start + target_frames
        segment = all_frames[start:end]
        
        # Pad with the last frame if segment is shorter
        while len(segment) < target_frames:
            segment.append(segment[-1])
        segments.append(np.array(segment))
    
    return segments
```

```python
import cv2

def extract_video_features(video_path, segment_duration=2, step_size=1, fps=5):
    """
    Extract features from the video, segment it, and apply the models to get deep and geometric features.

    :param video_path: Path to the video file
    :param segment_duration: Duration of each segment in seconds
    :param step_size: Step size for moving window segmentation
    :param fps: Frames per second to downsample to
    :return: Tuple (deep_features, geometric_features)
    """
    # Initialize video capture and get video details
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_fps = cap.get(cv2.CAP_PROP_FPS)

    # Downsample the video to 5 FPS (or any required FPS)
    frame_interval = int(video_fps / fps)
    
    # Extract frames in segments
    video_frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        # Downsampling frames by frame_interval
        if len(video_frames) % frame_interval == 0:
            video_frames.append(frame)
    
    cap.release()

    # Prepare for segmentation (2 seconds long segments)
    segments = []
    segment_size = segment_duration * fps  # Number of frames per segment
    for start_frame in range(0, len(video_frames) - segment_size + 1, step_size * fps):
        segments.append(video_frames[start_frame:start_frame + segment_size])
    
    return segments
```

```python
import numpy as np
from mediapipe import solutions as mp_solutions

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp_solutions.face_mesh.FaceMesh(min_detection_confidence=0.5, min_tracking_confidence=0.5)

def extract_facial_landmarks(segment):
    """
    Extract facial landmarks from the video segment using MediaPipe.
    :param segment: List of frames in the segment
    :return: List of facial landmarks for each frame
    """
    face_landmarks_list = []
    with mp_face_mesh.FaceMesh(min_detection_confidence=0.5, min_tracking_confidence=0.5) as face_mesh:
        for frame in segment:
            # Convert frame to RGB
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = face_mesh.process(frame_rgb)
            
            if results.multi_face_landmarks:
                face_landmarks = results.multi_face_landmarks[0]  # Assuming one face per frame
                landmarks = np.array([[lm.x, lm.y, lm.z] for lm in face_landmarks.landmark]).flatten()
                face_landmarks_list.append(landmarks)
            else:
                face_landmarks_list.append(np.zeros(468*3))  # In case no face is detected
    return np.array(face_landmarks_list)
```

```python
def extract_deep_features(segment, model):
    """
    Extract deep features from the segment using the deep model (e.g., Emo-AffectNet).
    :param segment: List of frames in the segment
    :param model: The pre-trained deep model for emotional features
    :return: Extracted deep features for the segment
    """
    # Prepare the segment (feature extraction should align with model input)
    features = torch.tensor(segment).unsqueeze(0)  # Add batch dimension if needed
    with torch.no_grad():
        deep_features = model(features)
    return deep_features.cpu().numpy()
```

```python
def extract_geometric_features(landmarks, model):
    """
    Extract geometric features using the hand-crafted model.
    :param landmarks: Facial landmarks from the segment
    :param model: The pre-trained model for geometric features
    :return: Extracted geometric features
    """
    landmarks_tensor = torch.tensor(landmarks).unsqueeze(0).float()  # Add batch dimension
    with torch.no_grad():
        geo_features = model(landmarks_tensor)
    return geo_features.cpu().numpy()
```

```python
def extract_video_frames(video_path, segment_length=5):
    """
    Extract frames from the video. We'll assume 1 frame per second.
    :param video_path: Path to the video file
    :param segment_length: Length of each segment in seconds
    :return: List of frames for each segment
    """
    video_frames = []
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        frame_count += 1
        # Assuming the frame rate is consistent, take one frame per second
        if frame_count % int(fps) == 0:
            video_frames.append(frame)

        # Stop if we've processed all frames
        if frame_count > total_frames:
            break

    cap.release()
    return np.array(video_frames)  # Returning frames as numpy array
```

```python
import cv2
import os

def extract_video_frames(video_path, frame_rate=30):
    """Extract frames from a video file at the given frame rate."""
    cap = cv2.VideoCapture(video_path)
    frames = []
    
    # Ensure the video is opened
    if not cap.isOpened():
        print(f"Error opening video file: {video_path}")
        return []

    # Extract frames at a fixed frame rate
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    for i in range(frame_count):
        ret, frame = cap.read()
        if not ret:
            break
        if i % frame_rate == 0:  # Adjust frame_rate as needed
            frames.append(frame)
    cap.release()
    return frames
```
