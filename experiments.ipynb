```python
import pandas as pd

# Load the feature files
audio_df = pd.read_csv('/kaggle/input/first-impression-v2-a-v-t-features/UPDATED/audio_features.csv')
text_deep_df = pd.read_csv('/kaggle/input/first-impression-v2-a-v-t-features/UPDATED/text_deep_features.csv')
text_handcrafted_df = pd.read_csv('/kaggle/input/first-impression-v2-a-v-t-features/UPDATED/text_handcrafted_features.csv')
video_df = pd.read_csv('/kaggle/input/first-impression-v2-a-v-t-features/UPDATED/video_features.csv')

# Check a few rows to understand the structure

```

```python
text_deep_df['Filename'] = text_deep_df['Filename'].str.removesuffix(".mp4")
text_handcrafted_df['Filename'] = text_handcrafted_df['Filename'].str.removesuffix(".mp4")
```

```python
# Check a few rows to understand the structure
print("\nAUDIO DEEP AND HANCRAFTED FEATURES: \n",audio_df.head())
print("\nTEXT DEEP FEATURES: \n",text_deep_df.head())
print("\nTEXT DEEP HANCRAFTED FEATURES: \n",text_handcrafted_df.head())
print("\nVIDEO DEEP AND HANCRAFTED FEATURES: \n",video_df.head())
```

```python
# audio_features = audio_df
# video_features = video_df
# Select relevant feature columns (excluding the identifier and target columns)
audio_features = audio_df.drop(['Audio_ID' , 'Segment_ID','Video_File', 'Ethnicity', 'Gender', 'AgeGroup','extraversion',	'neuroticism',	'agreeableness',	'conscientiousness'	,'interview'	,'openness'], axis=1)
video_features = video_df.drop([ 'Video_ID' , 'Segment_ID', 'Video_File', 'Ethnicity', 'Gender', 'AgeGroup','extraversion',	'neuroticism',	'agreeableness',	'conscientiousness',	'interview',	'openness'], axis=1)
text_deep_features = text_deep_df.drop(['Filename','Video_File', 'Ethnicity', 'Gender', 'AgeGroup','extraversion',	'neuroticism',	'agreeableness',	'conscientiousness'	,'interview'	,'openness'], axis=1)
text_hc_features = text_handcrafted_df.drop([ 'Filename' ,'Video_File', 'Ethnicity', 'Gender', 'AgeGroup','extraversion',	'neuroticism',	'agreeableness',	'conscientiousness'	,'interview'	,'openness'], axis=1)

```

```python
import pandas as pd

# # Load the audio and video features
# audio_features = audio_df
# video_features = video_df

# Splitting the columns into deep and handcrafted for audio
audio_deep_columns = [col for col in audio_features.columns if col.startswith('Deep')]
audio_hc_columns =  [col for col in audio_features.columns if col.startswith('Hc')]

audio_deep_features = audio_features[audio_deep_columns]
audio_hc_features = audio_features[audio_hc_columns]

# Renaming the audio deep columns
audio_deep_features.columns = [f'Audio_deep_{i}' for i in range(--audio_deep_features.shape[1])]

# Renaming the audio handcrafted columns
audio_hc_features.columns = [f'Audio_hc_{i}' for i in range(audio_hc_features.shape[1])]

# Splitting the columns into deep and handcrafted for video
video_deep_columns =  [col for col in video_features.columns if col.startswith('Deep')]
video_hc_columns =   [col for col in video_features.columns if col.startswith('Hc')]

video_deep_features = video_features[video_deep_columns]
video_hc_features = video_features[video_hc_columns]

# Renaming the video deep columns
video_deep_features.columns = [f'Video_deep_{i}' for i in range(video_deep_features.shape[1])]

# Renaming the video handcrafted columns
video_hc_features.columns = [f'Video_hc_{i}' for i in range(video_hc_features.shape[1])]

# Let's preview the splits (only first few rows)
print("Audio Deep Features Head:")
audio_deep_features = pd.concat([audio_df[['Audio_ID','Segment_ID']],audio_deep_features],axis=1)
audio_deep_features=audio_deep_features.join(audio_df[['openness','conscientiousness','extraversion','agreeableness','neuroticism','interview','Gender','Ethnicity','AgeGroup']])
print(audio_deep_features.head())

print("\nAudio Handcrafted Features Head:")
audio_hc_features = pd.concat([audio_df[['Audio_ID','Segment_ID']],audio_hc_features],axis=1)
audio_hc_features=audio_hc_features.join(audio_df[['openness','conscientiousness','extraversion','agreeableness','neuroticism','interview','Gender','Ethnicity','AgeGroup']])
print(audio_hc_features.head())

print("\nVideo Deep Features Head:")
video_deep_features = pd.concat([video_df[['Video_ID','Segment_ID']],video_deep_features],axis=1)

video_deep_features=video_deep_features.join(video_df[['openness','conscientiousness','extraversion','agreeableness','neuroticism','interview','Gender','Ethnicity','AgeGroup']])
print(video_deep_features.head())

print("\nVideo Handcrafted Features Head:")
video_hc_features = pd.concat([video_df[['Video_ID','Segment_ID']],video_hc_features],axis=1)

video_hc_features=video_hc_features.join(video_df[['openness','conscientiousness','extraversion','agreeableness','neuroticism','interview','Gender','Ethnicity','AgeGroup']])
print(video_hc_features.head())

```
